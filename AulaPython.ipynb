```python
# Dicionário de renomeação
novos_nomes = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

# Aplicando renomeação
df.rename(columns=novos_nomes, inplace=True)

# Verificando resultado
df.head()
```

```python
df.columns
```

```python
df["senioridade"].value_counts()
```

```python
# Dicionário de mapeamento para traduzir os níveis de experiência
senioridade = {
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
}

# Aplicar o mapeamento à coluna 'nivel_experiencia'
df['senioridade'] = df['senioridade'].replace(senioridade)

# Verificar as novas categorias traduzidas
print(df['senioridade'].value_counts())
```

```python
# Dicionário de mapeamento para traduzir os tipos de contrato
contrato = {
    'FT': 'Tempo Integral',
    'PT': 'Tempo Parcial',
    'CT': 'Contrato',
    'FL': 'Freelancer'
}


df['contrato'] = df['contrato'].replace(contrato)
print(df['contrato'].value_counts())
```

```python
# Dicionário de mapeamento para traduzir os tamanhos de empresa
tamanho_empresa = {
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
}

# Aplicar o mapeamento à coluna 'tamanho_empresa'
df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
print(df['tamanho_empresa'].value_counts())
```

```python
# Dicionário de mapeamento para traduzir as categorias da coluna 'remoto'
remoto = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

# Aplicar o mapeamento à coluna 'remoto'
df['remoto'] = df['remoto'].replace(remoto)
print(df['remoto'].value_counts())
```

```python
import numpy as np
#Criação de um dataframe de teste
df_salarios = pd.DataFrame({
   'nome': ["Ana", "Bruno", "Carlos", "Daniela", "Vanessa"],
   'salario': [4000, np.nan, 5000, np.nan, 100000]
})

#Calcula a média salarial e substitui os valores nulos pela média, arredondando os valores.
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
#Calcula a mediana salarial e substitui os valores nulos pela mediana.
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"],
    "Temperaturas": [30, np.nan, np.nan, 27, 24]
})

df_temperaturas["preenchido_ffill"] = df_temperaturas["Temperaturas"].ffill()
df_temperaturas
```
